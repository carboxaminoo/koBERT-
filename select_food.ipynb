```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/AI프로젝트 입문/food_feeling.csv')
```

```python
def find_ingredient(df,Feel):
  sence=df[df['감정'].isin([Feel])]
  sampling=sence.sample(n =1)
  ingredient = sampling.iloc[0,1]
  element = sampling.iloc[0,2]
  return ingredient, element
```

```python

Feel = 6

#공포 놀람
if Feel -2 <0 :
  Feeling = '진정'
  print(find_ingredient(df,Feeling))

#분노 혐오
elif Feel ==2 or 6:
  Feeling = '스트레스'
  a,b=find_ingredient(df,Feeling)
  print(a+'에는 '+b+' 성분이 들어있습니다')

#슬픔
elif Feel == 3:
  Feeling = '우울'
  print(find_ingredient(df,Feeling))

#행복
elif Feel == 5:
  print("오늘은 가장 좋아하는 음식을 드시는건 어떤가요?")

#중립
else: 
  print("추천할 음식이 없습니다.")
```
